## Function

```python
def cn_close_price_dfer(digital_asset):
    
    # Date information 
    startdate = datetime(2018,1,1)
    enddate = datetime.today() 
    
    # API PULL to get Data for coins
    adj_close = web.DataReader(digital_asset, 'yahoo',startdate, enddate)
    adj_close_df = adj_close.filter(like="Adj Close")
  


    
    return(adj_close_df)
```

## Code

```python
string = "-USD"
stock_list = input("Enter the ticker you want to explore:")

stock_df = cn_close_price_dfer(stock_list)
```

```python
display(stock_df.head())
print("\n")
display(stock_df.tail())
```

```python
stock_df["Returns"]=stock_df["Adj Close"].pct_change()
# stock_df["Standard Dev"] = stock_df["Returns"].rolling(window=2).std()
# stock_df["Sharpe"]= stock_df["Returns"]/stock_df["Standard Dev"]
stock_df
```

```python
stock = yf.Ticker(stock_list)
stock.major_holders
```

```python
short_window = 7
long_window = 90

stock_df['sma_fast'] = stock_df['Adj Close'].rolling(window=short_window).mean()
stock_df['sma_slow'] = stock_df['Adj Close'].rolling(window=long_window).mean()
```

```python
stock_df.dropna()
```

```python
X = stock_df[['sma_fast', 'sma_slow']].shift().dropna().copy()
```

```python
X.head()
```

```python
stock_df['signal'] = 0.0
stock_df.loc[(stock_df['Returns'] >= 0), 'signal'] = 1
stock_df.loc[(stock_df['Returns'] < 0), 'signal'] = -1              
```

```python
y = stock_df['signal'].copy()
```

```python
training_begin = X.index.min()
print(training_begin)

training_end = X.index.min() + DateOffset(months=6)
print(training_end)
```

```python
# Creating the X_train and y_train DataFrames

X_train = X.loc[training_begin:training_end]
y_train = y.loc[training_begin:training_end]

X_train.head()
```

```python
# Creating the X-test and y_test DataFrames

X_test = X.loc[training_end:]
y_test = y.loc[training_end:]

X_test.head()
```

```python
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Test machine model SVM (support vector machine learning method)

svm_model = svm.SVC()
svm_model = svm_model.fit(X_train_scaled, y_train)
training_signal_predictions = svm_model.predict(X_train_scaled)

training_signal_predictions[:10]

```

```python
# Evaluation SVM model using classification report

training_report = classification_report(y_train, training_signal_predictions)
print(training_report)
```

```python
# Backtesting the SVM model

testing_signal_predictions = svm_model.predict(X_test_scaled)
testing_report = classification_report(y_test, testing_signal_predictions)
print(testing_report)
```

```python
# Comparing actual and predicted returns
predictions_df = pd.DataFrame(index=X_test.index)

predictions_df["predicted_signal"] = testing_signal_predictions

predictions_df["Returns"] = stock_df["Returns"]

predictions_df["trading_algorithm_returns"] = (
    predictions_df["Returns"] * predictions_df["predicted_signal"]
)

# Review the DataFrame
predictions_df.head()
```
